```python

estaciones=3
u= 100
m= 5
Q= 50
archivo = open("BRT-C1-S3-GENERADO.txt", "w") 



rutas=[]
rutas.append( (1,1,1) ) 
rutas.append( (0,1,1) ) 
rutas.append( (1,0,1) ) 
rutas.append( (1,1,0) ) 
   
    
    
    
nodos=[]
for i in range(1,estaciones+1):
    nodos.append(i)
    

#rutas=[(a,b,c,d) for a in range(2)for b in range(2)for c in range(2)for d in range(2)]

archivo.write("Slabel:[\n")
for i in range(len(nodos)):
    archivo.write("(")
    archivo.write(str(nodos[i]))
    archivo.write(")'S")
    archivo.write(str(nodos[i]))
    archivo.write("'"+"\n")
archivo.write("]\n")
archivo.write("\n")
archivo.write("Dlabel:[\n")

archivo.write("]\n")
archivo.write("\n")




c ={}
for i in nodos:
    for j in nodos:
        if( i != j ):
            c[ (i,j) ] = abs( i - j ) * 10 - abs( i - j ) + 1 


archivo.write("cost:[\n")
for name,dict_ in c.items():
    archivo.write(str(name))
    archivo.write(str(dict_)+"\n")
archivo.write("]\n")
archivo.write("\n")

archivo.write("u:"+str(u)+"\n")
archivo.write("m:"+str(m)+"\n")
archivo.write("Q:"+str(Q)+"\n")
archivo.write("\n")


stat={}
for i in nodos:
    stat[(i)]=i



b={}    
for i in nodos:
    for j in nodos:
        if( i != j ):
            b[ (i,j) ] = 50




Ar={}

for i in range(len(rutas)):
    aux=[]
    for j in range(estaciones):
        if(rutas[i][j]):
            aux.append(j+1)
    #print(aux)
    nodos.append(len(nodos)+1)
    stat[(len(nodos))]=aux[0]
    for j in range(1,len(aux)):
        nodos.append(len(nodos)+1)
        stat[(len(nodos))]=aux[j]

        Ar[(len(nodos)-1,len(nodos),i+1)]=1
        Ar[(len(nodos),len(nodos)-1,i+1)]=1
As={}

for i in range(len(nodos)):
    for j in range(len(nodos)):
        if(i!=j):
            if(stat[(nodos[i])]==stat[(nodos[j])]):
                As[((nodos[i]),(nodos[j]),stat[(nodos[j])])]=1


archivo.write("stat:[\n")
for name,dict_ in stat.items():
    archivo.write("(")   
    archivo.write(str(name))
    archivo.write(")")
    archivo.write(str(dict_)+"\n")
archivo.write("]\n")
archivo.write("\n")
  
    
archivo.write("b:[\n")
for name,dict_ in b.items():  
    archivo.write(str(name))
    archivo.write(str(dict_)+"\n")
archivo.write("]\n")
archivo.write("\n")    
   
archivo.write("Ar:[\n")
for name,dict_ in Ar.items():  
    archivo.write(str(name))
    archivo.write(str(dict_)+"\n")
archivo.write("]\n")
archivo.write("\n")     

archivo.write("As:[\n")
for name,dict_ in As.items():  
    archivo.write(str(name))
    archivo.write(str(dict_)+"\n")
archivo.write("]\n")
archivo.write("\n")   

archivo.close()
```
