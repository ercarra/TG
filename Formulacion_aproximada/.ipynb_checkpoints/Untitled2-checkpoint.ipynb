```python

fic = open("datos4.txt", "r")
lines = []
for line in fic:
    lines.append(line)
import re 

import numpy as np

numeros = re.compile(r'[0-9]+\.*[0-9]*')
i=0
k=int(lines[i])
i=i+1
result = numeros.findall(lines[i])
customers=int(result[0])
tams=[]
for j in range(k):
    tams.append(int(result[j+1]))



i=i+2
f=[] 

F=np.zeros((tams[j]))
print(tams)
for j in range(k):
    F=np.zeros((tams[j]))
    for s in range(tams[j]):
        result = numeros.findall(lines[i])
        F[s]=float(result[0])
        i=i+1
    f.append(F)    

i=i+1

C=[]
c=np.zeros((customers,tams[0]))
C.append(c)

for j in range(k-1):
    c=np.zeros((tams[j],tams[j+1]))
    C.append(c)
    
for j in range(customers):
     result = numeros.findall(lines[i])
     i=i+1
     for s in range(tams[0]):
        C[0][j][s]=result[s]
i=i+1

for j in range(k-1):
    for s in range(tams[j]):
        result = numeros.findall(lines[i])
        i=i+1
        for t in range(tams[j+1]):
            C[j+1][s][t]=result[t]
    i=i+1
    
#for j in range(k):
 #   tams.append(result[j+1])
#print (tams)
from docplex.mp.model import Model
mdl= Model("modelo")


arcosv=[(i,j) for i in range(customers) for j in range(tams[0])]
v=mdl.binary_var_dict(arcosv,name="v")

arcosz=[(i,r,a,b) for i in range(customers) for r in range(1,k) for a in range(tams[r-1]) for b in range(tams[r]) ]
z=mdl.binary_var_dict(arcosz,name="z")

arcosy=[(r,s) for r in range(k) for s in range(tams[r])]
y=mdl.binary_var_dict(arcosy,name="y")
for i in range(customers):
    mdl.add_constraint(mdl.sum( v[(i,j)] for j in range(tams[0]) )==1 )
mdl.minimize( mdl.sum( C[0][i][j]*v[(i,j)] for i in range(customers) for j in range(tams[0]))+mdl.sum( z[i,r,a,b]*C[r][a][b]  for i in range(customers) for r in range(1,k) for a in range(tams[r-1]) for b in range(tams[r]) )+  mdl.sum( f[j][r]*y[(j,r)] for j in range(k) for r in range(tams[j]) ))
for i in range(customers):
    for j in range(tams[0]):
        mdl.add_constraint(  mdl.sum( z[(i,1,j,b)] for b in range(tams[1]) )==v[(i,j)] )
for i in range(customers):
    for j in range(tams[0]):
        mdl.add_constraint(v[(i,j)]<=y[(0,j)])  
for i in range(customers):
    for r in range(2,k):
        for a in range(tams[r-1]):
            mdl.add_constraint(  mdl.sum( z[(i,r,a,b)] for b in range(tams[r]) )==mdl.sum( z[(i,r-1,b,a)] for b in range(tams[r-2]) ) )        
for i in range(customers):
        for r in range(2,k+1):
            for b in range(tams[r-1]):
                mdl.add_constraint(mdl.sum( z[(i,r-1,a,b)] for a in range(tams[r-2]) )<=y[(r-1,b)] )
solucion=mdl.solve(log_output=True)  
archivo = open("solucion4.txt", "w") 
archivo.write(str(solucion)) 
archivo.close()
fic.close()

####---------------------------------------------------------------------------------

```
