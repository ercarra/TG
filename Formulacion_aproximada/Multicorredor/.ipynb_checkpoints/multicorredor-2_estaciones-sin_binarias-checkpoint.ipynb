
$Se \  tiene \ 2 \ corredores \ C_{1} y \ C_{2}, \  vectores \  ordenados , \ tal \ que \  C_{1} \cup C_{2}= N=\{1,...,n\}$
\
\
$  K_{1} \  todas \  las \  rutas \  posibles \ en \ el \ corredor \ C_{1} \  y \  K_{2} \  todas \  las \  rutas  \ posibles  \ en  \ el \  corredor  \ C_{2}\  $
\
\
$sea \  c_{ij} \  el \  tiempo \  requerido \  para  \ viajar \  entre \  la  \ estacion \  i \  y  \ j$
\
\
$\bar{c}_{ij}= \frac{c_{ij}}{2}  \ si  \ i \ y  \ j \in C_{1}\cap C_{2}$
\
$ \bar{c}_{ij}=c_{ij}\  e.o.c$
\
\
$\textbf{Problema Primal}$ 
\
\
$\displaystyle min \ \ \sum_{o \in N}\sum_{d \in N}\sum_{i \in C_{1}}\sum_{j \in C_{1}}\bar{c}_{ij}x_{ij}^{od}+
 \ \sum_{o \in N}\sum_{d \in N}\sum_{i \in C_{2}}\sum_{j \in C_{2}}\bar{c}_{ij}x_{ij}^{od}+
\alpha_{1} \sum_{k \in K_{1}}(f_{k})+\alpha_{2} \sum_{k \in K_{2}}(f_{k}) $
\
\
$Restricciones$
\
\
$\displaystyle\sum_{j \in C_{1} \cup C_{2}=N} x_{ij}^{od} -\sum_{j \in C_{1} \cup C_{2}=N} x_{ji}^{od}=b_{i}^{od} \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \forall  i \in C_{1} \cap C_{2}=, \ \forall  o \in N, \ \forall  d \in N$ 
$\displaystyle  \ \ \ \   \sum_{j \in C_{1} } x_{ij}^{od} -\sum_{j \in C_{1} } x_{ji}^{od}=b_{i}^{od} \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  \ \ \  \ \ \  \ \ \ \ \ \   \ \ \  \forall  i \in C_{1} \setminus C_{2}=, \ \forall  o \in N, \ \forall  d \in N$ 
$\displaystyle  \ \ \ \   \sum_{j \in C_{2} } x_{ij}^{od} -\sum_{j \in C_{2} } x_{ji}^{od}=b_{i}^{od} \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  \ \ \  \ \ \  \ \ \ \ \ \   \ \ \  \forall  i \in C_{2} \setminus C_{1}=, \ \forall  o \in N, \ \forall  d \in N$ 
\
\
$\displaystyle  \ \ \ \   \sum_{o \in N}\sum_{d \in N} x_{ij}^{od} \leq \sum_{k \in K_{1}}u*r_{kij}*f_{k} \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \forall  i \in C_{1}, \ \forall  j \in C_{1} $
\
$\displaystyle  \ \ \ \   \sum_{o \in N}\sum_{d \in N} x_{ij}^{od} \leq \sum_{k \in K_{2}}u*r_{kij}*f_{k} \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \forall  i \in C_{2}, \ \forall  j \in C_{2} $
\
\
$\displaystyle  \ \ \ \  \sum_{k \in K_{1}\cup K_{2}} f_{k} \leq Q$
\
\
$ la \  generación \  de \  rutas \  se  \ deduce \  igual \  que \  el \  problema  \ sin  \ multiples \  estaciones \  solo \  que  \ ahora \  existen \  2 \  tipos  \ de \  restricciones \  de  \ capacidad( \ segun  \ el \  corredor), \  por  \ lo \  que \  se  \ producen \  los  \ siguientes \  2 \  subproblemas $
\
\
$\textbf{Generación de rutas estacion } C_{1} $ 
\
\
$ \displaystyle max \ \ \lambda-\alpha_{1}- \sum_{i \in C_{1}}\sum_{j \in C_{1}}r_{ij}\beta_{ij}u$
\
\
$Restricciones$
\
\
$ \displaystyle\sum_{j \in C_{1}\setminus \ C_{11}}r_{C_{11}j} \leq s_{1}\ \  \ \  \ \ \  \ \ \  \ \ \ \ \ \ \ \ \  \ \ \  \ \ \ \ \ \ ( \  C_{1j} \  representa  \ el  \ j-avo \  elemento \  de \  C_{1}) $
\
$\displaystyle\sum_{j=i+1}^{ \lvert C_{1}\rvert}r_{iC_{1j}} \leq s_{i}+ \sum_{k=1}^{i-1}r_{C_{1k}i}, \ \  \ \ \  \ \ \  \ \ \ \ \ \ \ \forall i \in C_{1}\setminus \ C_{11} $
\
$\displaystyle\sum_{i=1}^{\lvert C_{1}\rvert} s_{i} \leq 1 $
\
$r_{ij}=r_{ji}\ \  \ \ \  \ \  \ \  \ \ \  \ \ \  \ \ \  \ \ \  \ \ \  \ \ \ \  \ \ \ \  \ \ \  \ \ \  \ \ \ \ \ \ \ \forall i \in C_{1}, \forall j \in C_{1}$
\
$r_{ii}=0, \ \ \  \ \ \ \   \ \  \ \ \  \ \ \  \ \ \  \ \ \  \ \ \  \ \ \ \  \ \ \ \  \ \ \  \ \ \  \ \ \ \ \ \ \  \forall i \in C_{1}$
\
\
\
$r_{ij} \in \{0,1 \}\ \ \  \ \ \ \ \ \ \  \  \  \ \ \  \ \ \  \ \ \ \  \ \ \ \  \ \ \  \ \ \  \ \ \ \ \ \ \ \forall i \in C_{1}\ ,j \in C_{1}$
\

\

$\displaystyle\sum_{i=1}^{\lvert C_{2}\rvert} s_{i} \leq 1 $
